```python
diary.drop(list(diary.columns)[11:],axis=1,inplace=True)
diary.dropna(how='all',inplace=True)
diary.tail()
```

```python
# fill in NaN, delete time and subtype
diary.Date.fillna(method="ffill",inplace=True)
diary.tail()
```

```python
diary.drop(['Time','Subtype'],axis=1,inplace=True)
diary.tail()
```

```python
# a dataframe of places
placeSer=diary['Place'].dropna().value_counts()
placeSer
```

```python
place_df=pd.DataFrame(placeSer)
place_df.rename(columns={'Place':'freq'},inplace=True)
place_df.index.name='PlaceName'
place_df
```

```python
# a dataframe of PersonName
diary_person=diary.drop(['Date','Type','Place'],axis=1).dropna(how='all')
diary_person
```

```python
name=pd.Series()
for idx,col in diary_person.iteritems():
    name=name.add(col.value_counts(),fill_value=0)
    ## element-wise add, fill_value for NaN
name_df=pd.DataFrame(name)
name_df.rename(columns={0:'謝蘭生'},inplace=True)
name_df
```

```python
name_df.to_csv("/Users/yumeng.zou/Downloads/name_test.csv")
```

```python
# a dataframe of relationship based on frequency
name_df.T
```

```python
people_df=name_df.join(name_df.T,how='outer')
people_df.fillna(0,inplace=True)
people_df
```

```python
diary_person.head()
```

```python
for idx,row in diary_person.iterrows():
    r=row[pd.notnull(row)]
    if len(r)>1:
        for name1 in r:
            for name2 in r:
                if name2!=name1:
                    people_df[name1][name2]+=1

people_df
```

```python
export=people_df.replace('0',np.nan)
export.to_csv("/Users/yumeng.zou/Downloads/people_test.csv")
```
