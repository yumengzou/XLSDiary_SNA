```python
JQ24=pd.read_csv("JQ24 diary.csv")
JQ25=pd.read_csv("JQ25 diary.csv")
DG1=pd.read_csv("DG1 diary.csv")
```

```python
diary=pd.concat([JQ24,JQ25,DG1])
diary.dropna(axis=[0,1],how='all',inplace=True)
diary.head()
```

```python
ppl_event=diary[['Type','Participants']].dropna(subset=['Participants'])
event_iter=0
event_idx=[]
for boo in ppl_event['Type'].notnull():
    if boo:
        event_iter+=1
    event_idx.append(event_iter)
    
ppl_event["Event"]=event_idx
ppl_event.drop("Type",axis=1,inplace=True)
#ppl_event=ppl_event.reset_index().drop("index",axis=1)
ppl_event.head()
```

```python
evtXie=pd.DataFrame({"Event":np.arange(1,event_iter+1),"Participants":"謝蘭生"})
evtXie.head()
```

```python
ppl_event=ppl_event.append(evtXie,ignore_index=True)
#ppl_event=ppl_event.reset_index().drop("index",axis=1)
ppl_event
```

```python
el=ppl_event.merge(ppl_event,on="Event")
el
```

```python
el.rename(columns={"Participants_x":"Source","Participants_y":"Target"},inplace=True)
```

```python
el=el.query("Source<Target")
el
```

```python
len(el.query("Participants_x<Participants_y")),len(el.query("Participants_y<Participants_x"))
```

```python
edgelist=el.query("Participants_x<Participants_y").drop("Event",axis=1)
edgelist.columns=["Source","Target"]
edgelist["Weight"]=1
edgelist
```

```python
edgelist=edgelist.groupby(["Source","Target"]).sum().reset_index()
```

```python
G = nx.from_pandas_dataframe(edgelist, source='Source', target="Target", edge_attr='Weight')
```

```python
nx.write_graphml(G, "test.graphml",encoding="utf-8")
```
