```python
people.rename(columns={'0':"name"},inplace=True)
people.set_index("name",inplace=True)
people
```

```python
people.replace(0,np.nan,inplace=True)
people.head()
```

```python
name.head()
```

```python
## create source, target, weight from dataframe people
i=0
Source=[]
Target=[]
Weight=[]
for colName,ser in people.iteritems():
    prep=ser[:i][pd.notnull(ser)]
    for idx,value in prep.iteritems():
#        print colName,idx,value
        Source.append(colName)
        Target.append(idx)
        Weight.append(value)
    i+=1
#edge=pd.DataFrame({"Source":Source,"Target":Target,"Type":157*["Undirected"],"Id":np.arange(157),"Weight":Weight})
```

```python
pd.merge(pd.DataFrame(Source),name,how='left',left_on=0,right_on="Label")
```

```python
## turn string into index
source=pd.merge(pd.DataFrame(Source),name,how='left',left_on=0,right_on="Label")["Id"]
target=pd.merge(pd.DataFrame(Target),name,how='left',left_on=0,right_on="Label")["Id"]
```

```python
edge=pd.DataFrame({"Id":np.arange(len(Weight)),"Source":source,"Target":target,"Type":len(Weight)*["Undirected"],"Weight":Weight})
edge
```
