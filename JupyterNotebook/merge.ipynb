```python
diary=pd.read_csv("Diary.csv")
diary.tail()
```

```python
evt_plc=diary[['Event','Place']].dropna()
evt_plc
```

```python
evt_type=diary[['Event','Type']].dropna()
evt_type
```

```python
plc_type=pd.merge(evt_plc,evt_type,how='left').drop("Event",axis=1)
```

```python
plc_type['Freq']=1
plc_type.groupby(['Place','Type']).sum().reset_index().sort_values(by='Freq',ascending=False)
```
