```python
ppl=pd.DataFrame({"Label":geoac["Participants"].unique(),"Modularity Class":"People"})
ppl
```

```python
place=pd.DataFrame({"Label":geoac["Place"].unique(),"Modularity Class":"Place"})
place
```

```python
nodes=pd.DataFrame(ppl.append(place,ignore_index=True))
nodes
```

```python
nodes.reset_index(inplace=True)
nodes.rename(columns={"index":"Id"},inplace=True)
nodes
```

```python
nodes.to_csv("/Users/yumeng.zou/Downloads/geoNode.csv",index=False)
```

```python
pd.merge(geoac,nodes,how='left',left_on="Place",right_on="Label")
```

```python
target=pd.merge(geoac,nodes,how='left',left_on="Place",right_on="Label")["Id"]
source=pd.merge(geoac,nodes,how='left',left_on="Participants",right_on="Label")["Id"]
source
```

```python
edges=pd.DataFrame({"Id":np.arange(len(source)),"Source":source,"Target":target,"Weight":1})
edges
```

```python
edges=edges.groupby(["Source","Target"]).sum().reset_index()
```
