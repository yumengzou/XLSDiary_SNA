```python
def visit(ser):
    if ser['Type']=='Visiting' or ser['Type']=='Visited':
        if ser['Type']=='Visiting':
            ser['Direction']='to'
        else:
            ser['Direction']='from'
        ser['Type']='Visit'
    return ser
```

```python
JQ25[['Type','Direction']].apply(visit,axis=1)
```

```python
def art(ser):
    if ser['Type']=='Painting' or ser['Type']=='寫字':
        ser['Type']='Art'
    if ser['Type']=='writing_for':
        ser['Type']='Art'
        ser['Direction']='for'
    return ser
```

```python
JQ25[['Type','Direction']].apply(art,axis=1)
```

```python
def invite(ser):
    if ser['Type']=='Inviting' or ser['Type']=='Invited':
        if ser['Type']=='Inviting':
            ser['Direction']='to'
        else:
            ser['Direction']='from'
        ser['Type']='Invitation'
    return ser
```

```python
JQ25[['Type','Direction']].apply(invite,axis=1)
```

```python
def go(s):
    if s=='Go_places' or s=='go':
        s='Go'
    return s
```

```python
JQ25['Type'].map(go)
```

```python
def gift(ser):
    if ser['Type']=='Gift_received' or ser['Type']=='Gift_sent':
        if ser['Type']=='Gift_received':
            ser['Direction']='from'
        else:
            ser['Direction']='to'
        ser['Type']='Gift'
    return ser
```

```python
JQ25[['Type','Direction']].apply(gift,axis=1)
```

```python
def recreation(s):
    if s=='Theater' or s=='彈琴':
        s='Recreation'
    return s
```

```python
JQ25['Type'].map(recreation)
```

```python
def occasion(ser):
    l=['道喜','道壽','Celebrate','Send_off','拜年']
    if ser['Type'] in l:
        ser=['Occasion',ser['Type']]
    return ser
```

```python
JQ25[['Type','Direction']].apply(occasion,axis=1)
```

```python
def banquet(s):
    if s=='Meal' or s=='飲':
        s=='Banquet'
    return s
```

```python
JQ25['Type'].map(banquet)
```
