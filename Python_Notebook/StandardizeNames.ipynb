```python
dNames={}
for key in names.index.unique():
    dNames[key]={}
    last=pd.Series(names.ix[key]["LastName"])
    dNames[key]["LastName"]=last.dropna()[0]
    first=pd.Series(names.ix[key]["FirstName"])
    dNames[key]["FirstName"]=first.dropna()[0]
    other=list(pd.concat([
        pd.Series(names.FirstName[key]),
        pd.Series(names.OtherNames[key]),
        pd.Series(names.OtherNames1[key])]).dropna())
    dNames[key]["OtherNames"]=other

dNames
```

```python
dNames2={}
for stdnm in names.index.unique():
    key=tuple(pd.concat(
        [pd.Series(names.ix[stdnm,'FirstName']),
         pd.Series(names.ix[stdnm,'OtherNames']),
         pd.Series(names.ix[stdnm,'OtherNames1'])]).dropna())
    dNames2[key]=stdnm
    
dNames2
```
