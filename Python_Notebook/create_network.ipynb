```python
edges = projection.merge(projection, on="journal")
```

```python
edges = edges.query("Source<Target")
```

```python
G = nx.from_pandas_dataframe(edges, source='Source', target="Target", edge_attr='weight')
```

```python
nx.write_graphml(G, "/Users/mwalsh/Networks_Erlin/shortzeitunimodal.graphml", encoding="utf-8")
```

```python
bb = nx.betweenness_centrality(G)
nx.set_node_attributes(G, 'betweenness', bb)
```
